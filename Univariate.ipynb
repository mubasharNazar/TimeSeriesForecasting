```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv("./sales-of-shampoo-over-a-three-ye.csv")
```

```python
df = df.iloc[:len(df)-1]
```

```python
df
```

```python
df.Timestamp = pd.to_datetime(df.Month,format='%m-%d') 
```

```python
df.index = df.Timestamp 
```

```python
df = df.drop("Month", axis=1)
```

```python
df
```

```python
df.shape
```

```python
train=df[0:30] 
test=df[30:]
```

# Autoregressive integrated moving average (ARIMA) 

```python
import statsmodels.api as sm
fit1 = sm.tsa.statespace.SARIMAX(train).fit()
```

```python
results = test.copy()
forecast = fit1.forecast(steps=6)
```

```python
results['forecast']= [i for i in forecast]
```
